# FuzzyWzzy

```python
# pip install rapidfuzz unidecode pandas
import pandas as pd
from rapidfuzz import process, fuzz
from unidecode import unidecode

def normaliza(s):
    if pd.isna(s): return ""
    s = unidecode(str(s).lower().strip())
    s = "".join(ch for ch in s if ch.isalnum() or ch.isspace())
    return " ".join(s.split())
```

```python
# Exemplo
left  = pd.DataFrame({"id":[1,2,3], "nome":["Farmácia São José", "HOSP Moinhos  Vento", "Clinica Sta.  Luzia"]})
right = pd.DataFrame({"cod":[10,11,12], "razao_social":["Hospital Moinhos de Vento","Clinica Santa Luzia","Farmacia Sao Jose LTDA"]})
```

```python
df_cnes_resumo
```

```python
# Exemplo
left  = df_ipea
right = df_cnes_resumo 
```

```python
left["key"]  = left["nome_udh"].map(normaliza)
right["key"] = right["bairro"].map(normaliza) 
```

```python
# Para cada registro da esquerda, pega a melhor correspondência na direita
matches = left["key"].apply(
    lambda q: process.extractOne(q, right["key"], scorer=fuzz.WRatio, score_cutoff=85)  # ajuste o cutoff
)
```

```python
matches
```

```python
# extractOne retorna (texto_casado, score, index_da_direita)
left[["match_key","score","right_idx"]] = pd.DataFrame(
    [(m[0], m[1], m[2]) if m else (None, 0, pd.NA) for m in matches],
    index=left.index
)
```

```python
resultado = (left
    .merge(right, left_on="right_idx", right_index=True, how="left", suffixes=("_left","_right"))
    .drop(columns=["key_right"])
    .rename(columns={"key_left":"chave_normalizada"})
)
resultado
```
